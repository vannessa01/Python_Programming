```python
#No. 4
def string_processor(string, proses = 'first_char'):
    proses = proses.lower()
    if proses =='first_char':
        hasil = string[0]
    elif proses == 'last_char':
        hasil = string[-1]
    elif proses == 'num_char':
        hasil = len(string)
    elif proses == 'num_vowels':
        num_v = 0
        for i in string:
            if i in ['a','i', 'u', 'e', 'o']:
                num_v += 1
        hasil = num_v
    else: hasil = 'periksa kembali'
    return string, hasil

```

```python
#No. 6
df1 = pd.read_excel('superstore.xls')
```

```python
df1['Year'] = df1['Order Date'].dt.year
df1.head(5)
```

```python
def salesdanprofit(region, tahun = 2017):
    filter = df1[(df1.Region == region) &
               (df1.Year == tahun)].reset_index()
    result = filter.groupby(['Region', 'Year']).agg({'Sales' : 'sum', 'Profit' : 'sum'})
    return result
```

```python
print(salesdanprofit('South', 2016))
print('-'*36)
print(salesdanprofit('West'))
```

```python
#No. 7
df2 = pd.read_excel('superstore.xls')
```

```python
df2['Year'] = df2['Order Date'].dt.year
df2.head()
```

```python
grb = df2.groupby(['Year', 'Customer ID', 'Customer Name']) ['Sales'].sum().reset_index()
grb.head()
```

```python
quantils = grb.groupby(['Year'])['Sales'].describe(percentiles=[.25, .5, .75]).reset_index()
quantils
```

```python
grb = grb.merge(quantils[['Year', '25%', '50%', '75%']], on = 'Year', how = 'inner')
grb.head()
```

```python
def segment(x):
    try:
        if x['Sales'] <= x['50%']:
            return 'Low'
        elif x['Sales'] <= x['75%']:
            return 'Medium'
        else:
            return 'High'
    except:
        pass
```

```python
grb['Segment'] = grb.apply(lambda x: segment(x), axis=1)
grb.head(5)
```

```python
#data 2017
grb[grb['Year'] == 2017].groupby(['Year', 'Segment']).agg({'Customer ID' : 'count'})
```

```python
#No 8
def perhitungan_lingkaran(r, metode):
    '''
    r adalah integer
    r adalah jari-jari lingkaran
    dalam fungsi perhitungan lingkaran terdapat 3 metode
    metode pertama menghitung diameter lingkaran dengan mengalikan jari-jari dengan 2
    metode kedua menghitung luas lingkaran dengan mengalikan phi(3.14) dengan jari-jari dengan jari-jari
    metode ketiga menghitung keliling lingkaran dengan mengalikan phi dengan 2 kali jari-jari
    
    '''
    try:
        if metode == 'diameter':
            hasil = 2*r
            return 'diameternya adalah', hasil
        elif metode == 'luas':
            hasil = 3.14*r*r
            return 'luasnya adalah', hasil
        elif metode == 'keliling':
            hasil = 3.14*2*r
            return 'kelilingnya adalah', hasil
        else:
            raise TypeError
    except TypeError:
            return("Invalid, rumus must be ('luas','keliling','diameter')")
        
```

```python
perhitungan_lingkaran(7, 'luas')
```

```python
perhitungan_lingkaran(7, 'diameter')
```
